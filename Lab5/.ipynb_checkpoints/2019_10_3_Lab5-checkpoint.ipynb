```python
def clean_seq(seq):
    """ Function to clean out any human made markings in the sequences
    
    i.e. removes ']', '[', and ' ' characters
    """
    return seq.replace(" ", "").replace("[", "").replace("]", "")

def findOBS(seq, OBS_seq):
    """ Function to find the given OBS sequence within the given sequence"""
    return (seq.index(OBS_seq), seq.index(OBS_seq) + len(OBS_seq) - 2)
# def addToDF(df, gate = None, i_1 = None, i_2 = None, seq = None, pos = (None, None)):
#     df = df.append(pd.DataFrame({"Gate":[gate], "Input-1":[i_1], "Input-2":[i_2], "sequence":[seq], "Start/End":[pos]}))
```

```python
def add_gate(df, original_seq = "", gate_name = None, obs_seqs = []): 
    """ Function to systematically add to the DataFrame DF:
            -> the sequence of: {the gate} x {all truth values} 
            
        :returns None"""
    
    truth_vals = (True, False)
    # print(truth_vals)
    num_inputs = len(obs_seqs) # The size of the truth tuples is equal to the number of OBS's
    truth_combos = itertools.product(truth_vals, repeat = num_inputs)
    # print(list(truth_combos))
    
    for truth_tuple in truth_combos: #For each of T/F tuple, (T, T, F, ...) in all 2^n possible tuples
        seq_i, start_stop = gen_seq(truth_tuple, obs_seqs, original_seq)
        # set_trace()
        print(seq_i)
        # Set the ith item of the dataframe to be the new entry (i.e. adding a new row)
        new_row = {"Gate":gate_name, 
                                     "Inputs":truth_tuple, 
                                     "sequence":seq_i, 
                                     "Start/End":start_stop}
        # print(new_row.columns)
        # set_trace()
        df.loc[len(df)] = new_row
       
                
        
def gen_seq(truth_tuple, obs_seqs, original_seq):
    """ Function to systematically generate the sequence
            given a truth tuple (T, T, F, T, ...)
            and a number of OBS sequences that correspond
            to each of the truth values
        
        > gen_seq( (T, F), "GGG", "AAAGGGAAGA")
            "AAAxxxAAGA", (3, 6)
            
            """
    def replace_with_X(sequence):
        """ Returns a sequence of the same length, but with 'x's """
        bound_OBS = "x"*len(sequence)
        return bound_OBS
    
    # Make sure there are enough OBS sequences for each T/F in the truth tuple
    assert len(truth_tuple) == len(obs_seqs) 
    
    seq_i = original_seq
    start_stop_pos = []
    for i in range(len(truth_tuple)):
        obs_i = obs_seqs[i]
        try:
            start_stop_i = (original_seq.index(obs_i), original_seq.index(obs_i) + len(obs_i))
        except ValueError:
            print("ERROR: Substring not found")
            return (None, None)
        start_stop_pos.append(start_stop_i) # Add the tuple to the list
        if truth_tuple[i]: # If the ith truth value in the tuple is TRUE:
            x_seq = replace_with_X(obs_i)
            seq_i = seq_i.replace(obs_i, x_seq) # The sequence with the ith OBS replaced with an equal number of 'x''s
            # print(seq_i)
        # seq_i[start_stop_pos[0]:start_stop_pos[1]] = x_seq # The sequence with the ith OBS replaced with an equal number of 'x''s
      
    return seq_i, start_stop_pos    
    
# Testing:
# print("Old df: ", df.columns)
# add_gate(df, original_seq = "AAAAAAGGAA", gate_name = "NOT", obs_seqs = ["GG"])
# df
# add_gate(df, original_seq = "AAAAAATTTATTAA", gate_name = "AND", obs_seqs = ["TTATT", "AAAA"])
# df
# add_gate(df, gate_name = "NOT", num_inputs = 1)

# Works!
```

```python
# Making a function to run a given piece of RNA:

def foldRNA(rna_seq, gate_i):
    """ Function to find folds in RNA sequence
    
    :returns string containing all folds"""
    
    
    proc = subprocess.run(["RNAfold"], input=bytes(rna_seq, 'ascii'), 
                              stdout=subprocess.PIPE, stderr=subprocess.PIPE, check=True)
#     print(proc.stderr.decode())
    os.system("mv rna.ps {new_name}.ps; ps2pdf {new_name}.ps {new_name}.pdf".format(new_name = gate_i))
    return proc.stdout.decode().split("\n")
    # print()
```

```python
seq_row = 3
truth_row = 2

def strTF(truth_tuple):
    s = ""
    for t in truth_tuple:
        s += str(t)
    return s
        
for row in df.iterrows():
    row = row[1]
    # print(row)
    seq = row["sequence"]
    t_f = row["Inputs"]
    gate_name = row["Gate"] 
    gate_i = gate_name + strTF(t_f)
    # print(gate_i)
    foldRNA(seq, gate_i)
```
