# Euler's method

```python
n = 100 # num steps for Euler's method
def dy_dt(t, y):
    """ Function that represents dy_dt = f(t, y)"""
    return t*t*t + y*y
y_0 = 0.0001

y = [y_0]
dt = 0.000001

for i in range(n):
    y_i = y[-1] + dy_dt(dt*i, y[-1])
    y.append(y_i)
    
print(y)
```
