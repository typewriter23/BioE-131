```python
import string
import matplotlib.pyplot as plt
```

```python
# First, read in the file using a python dictionary

textFile = "./romeoandjuliet.txt"

all_words = []
with open(textFile, "r") as book:
    long_str = book.read()
    
    # Clean newline chars:
    long_str = long_str.replace("\n", " ")
    
    # Make all words lowercase
    long_str = long_str.lower()
    
    # Clean punctuation 
    # (taken from: https://stackoverflow.com/questions/265960/best-way-to-strip-punctuation-from-a-string)
    long_str = long_str.translate(str.maketrans('', '', string.punctuation))
    
    # long_str = long_str.replace("  ", " ")
    
    all_words = long_str.split(" ")
    
    all_words[:] = [word for word in all_words if word != ""]
# all_words
```

```python
# Now, we have a list of words
# We now want to sort them based on frequency

def genDict(listA):
    """ Helper function to generate a dictionary of {[element]: occurance} pairs
            from a given list"""
    mappings = {}
    for elem in listA:
        if elem in mappings.keys():
            mappings[elem] += 1
        else:
            mappings[elem] = 1
            
    return mappings
```

```python
word_freqs = genDict(all_words)
print(word_freqs)
```
